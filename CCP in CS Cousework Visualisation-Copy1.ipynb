```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
percentage = [51, 39]
atti = ['Yes', 'No']
plt.figure(figsize=(5, 3), dpi=300)
plt.bar(atti,
        percentage,
        width=0.6,
        color=sns.color_palette(palette='deep')[0:2])
plt.title('Whether Teachers Find \nThis Behaviour Acceptable (%) (p.182)',
          fontsize=18)
plt.ylabel('Percentage', fontsize=12)
plt.xlabel('Attitude', fontsize=12)
plt.ylim(0, 100)
for x, y in enumerate(percentage):
    plt.text(x, y + 0.5, "%s" % y, ha='center')
plt.grid(axis='y', alpha=0.6)
plt.gca().set_axisbelow(True)
plt.show()
```

```python
plt.figure(figsize=(6, 10), dpi=300)
plt.pie(
    x=[38, 33, 24, 5],
    labels=['Acceptable', 'Collusion', 'Plagiarism', 'Other'],
    colors=sns.color_palette(palette='deep')[0:4],
    autopct='%.0f%%',
)
plt.legend(title="Attitude",
           loc="center left",
           fontsize=10,
           bbox_to_anchor=(1.1, 0, 0.5, 1))
plt.title('How Students Consider This Scenario (%) (p.182)',
          loc="left",
          fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10, 60), dpi=300)

from pylab import *

subplots_adjust(left=0.0, bottom=0.0, top=1, right=1)

plt.subplot(1, 2, 1)
plt.pie(
    x=[20, 80],
    labels=['Cheated', "Didn't"],
    colors=sns.color_palette(palette='hls'),
    autopct='%.0f%%',
)
plt.title(
    'Percentage distribution of Students whether \n Cheated on 30% of Course Materials or not (%) (p.186)',
    loc="center",
    fontsize=15)

plt.subplot(1, 2, 2)
plt.pie(
    x=[47, 53],
    labels=['Cheated over 30%', "Didn't"],
    colors=sns.color_palette(palette='hls'),
    autopct='%.0f%%',
)
plt.title(
    'Percentage distribution of Students that\n Failed the Course (%) (p.186)',
    loc="center",
    fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(6, 10), dpi=300)
plt.pie(
    x=[2, 75, 23],
    labels=[
        'Too many questions', 'Satisfied with the amount',
        'Preferred more questions'
    ],
    colors=sns.diverging_palette(240, 10, n=3),
    autopct='%.0f%%',
)
plt.legend(loc="lower left", fontsize=10, bbox_to_anchor=(1.1, 0, 0.5, 1))
plt.title(
    'What Students Thought about Responding 3-5 Questions per Lecture (%) (p.191)',
    loc="center",
    fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(6, 30), dpi=300)
labels = [
    "Useful, they helped clarify what's allowed",
    'Useful, although they only reinforced what I already knew',
    'I learned nothing from them', 'They were a waste of time',
    'They left me more confused'
]
x = [63, 25, 0, 0, 12]
plt.pie(x, colors=sns.color_palette("coolwarm", 5), autopct='%.0f%%')
plt.legend(labels,
           loc="lower left",
           fontsize=10,
           bbox_to_anchor=(1.1, 0, 0.5, 1))
plt.title(
    'How Useful Do Students Find the Lecture of \nUnderstanding the Acceptable Behaviour (%) (p.191)',
    loc="center",
    fontsize=18)
plt.show()
```
